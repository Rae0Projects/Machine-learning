```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Load the data from the CSV file
data = pd.read_csv('instagram_reach.csv')

# Function to convert time durations to minutes
def convert_to_minutes(time_str):
    if 'hour' in time_str:
        return int(re.search(r'\d+', time_str).group()) * 60
    elif 'minute' in time_str:
        return int(re.search(r'\d+', time_str).group())
    else:
        return None

# Convert the "Time since posted" column to numeric format (assuming it's in minutes)
data['Time since posted'] = data['Time since posted'].apply(convert_to_minutes)

# Drop rows where "Time since posted" couldn't be converted
data = data.dropna(subset=['Time since posted'])

# Set color palette
sns.set_palette("pastel")

# Scatter plot: Followers vs. Likes
plt.figure(figsize=(8, 6))
plt.scatter(data['Followers'].to_numpy(), data['Likes'].to_numpy(), alpha=0.5)
plt.xlabel('Followers')
plt.ylabel('Likes')
plt.title('Followers vs. Likes')
plt.grid(True)
plt.show()

# Line plot: Likes over Time Since Posted
plt.figure(figsize=(10, 6))
plt.plot(data['Time since posted'].to_numpy(), data['Likes'].to_numpy(), marker='o', linestyle='-')
plt.xlabel('Time Since Posted (Minutes)')
plt.ylabel('Likes')
plt.title('Likes over Time Since Posted')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Histogram: Distribution of Followers
plt.figure(figsize=(8, 6))
plt.hist(data['Followers'].to_numpy(), bins=20, edgecolor='black')
plt.xlabel('Followers')
plt.ylabel('Frequency')
plt.title('Distribution of Followers')
plt.grid(True)
plt.show()

# Set color palette
sns.set_palette("muted")

# Box plot: Spread and Outliers in Likes
plt.figure(figsize=(8, 6))
sns.boxplot(data['Likes'])
plt.ylabel('Likes')
plt.title('Box Plot of Likes')
plt.show()

# Pair plot: Relationship between all numerical columns
sns.pairplot(data[['Followers', 'Likes', 'Time since posted']])
plt.show()

# Additional Graphs
# Bar plot: Average Likes for different Captions
plt.figure(figsize=(10, 6))
sns.barplot(x='Caption', y='Likes', data=data)
plt.xlabel('Caption')
plt.ylabel('Average Likes')
plt.title('Average Likes for different Captions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar plot: Average Likes for different Hashtags
plt.figure(figsize=(10, 6))
sns.barplot(x='Hashtags', y='Likes', data=data)
plt.xlabel('Hashtags')
plt.ylabel('Average Likes')
plt.title('Average Likes for different Hashtags')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
